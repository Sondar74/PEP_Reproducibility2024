```python
# Runge-Kutta methods implementation
def runge_kutta(f, t_span, y0, h, A, b, c):
    t0, tn = t_span
    num_steps = int((tn - t0) / h)
    t_values = [t0]
    y_values = [y0]
    
    for _ in range(num_steps):
        t = t_values[-1]
        y = y_values[-1]
        
        k = [np.zeros_like(y) for _ in range(len(c))]
        
        for i in range(len(c)):
            ti = t + c[i] * h
            yi = y.copy()
            for j in range(i):
                yi += A[i][j] * h * k[j]
            k[i] = f(ti, yi)
        
        y_new = y.copy()
        for i in range(len(b)):
            y_new += b[i] * h * k[i]
        
        t_values.append(t + h)
        y_values.append(y_new)
    
    return t_values, y_values

# Useful constants
e = math.e
```

```python
# Convergence Test:

class CustomArray:
    def __init__(self):
        self.h = None
        self.sol_error = None
        self.sol_EOC = None
        self.energy_error = None
        self.energy_EOC = None
        
class EnergyArray:
    def _init_(self):
        self.final_error = None
        self.Energy_T = None
        self.Energy_exp = None

def EOC(e1, e2):
    eoc = np.log(e1/e2)/np.log(2)
    eoc = "{:.2e}".format(eoc)
    return eoc
    

def EOC_array(h1):
    # Create a table
    my_array = CustomArray()
    
    # Initial error to calculate EOC
    sol1 = solution(h1)
    h2 = h/2
    sol2 = solution(h2)
    
    error1 = sol1.final_error
    error2 = sol2.final_error
    
    # Error for energy
    energy_error1 = abs(sol1.Energy_T - sol1.Energy_exp)
    energy_error2 = abs(sol2.Energy_T - sol2.Energy_exp)
    
    # Fill the entries
    my_array.h = h1
    my_array.sol_EOC = EOC(error1,error2)
    my_array.energy_error = energy_error1
    my_array.energy_EOC = EOC(energy_error1, energy_error2)
    
    # Format
    error1 = "{:.2e}".format(error1)
    energy_error1 = "{:.2e}".format(energy_error1)
    
    my_array.sol_error = error1
    my_array.energy_error = energy_error1
    
    return my_array
    
    
# Analitic solution

def analytical(t):
    y =  np.log((np.exp((math.sqrt(math.e)+math.e)*t)*(math.sqrt(math.e)+math.e))/(math.sqrt(math.e)+np.exp((math.sqrt(math.e)+math.e)*t)))
    x =  np.log(math.e+np.exp(3/2)) - np.log(math.sqrt(math.e)+np.exp((math.sqrt(math.e)+math.e)*t))
    return [x, y]

# Set for Conserved exponential entropy
def f(t, y):
    u1, u2 = y
    u1_dot = -np.exp(u2)
    u2_dot = np.exp(u1)
    return np.array([u1_dot, u2_dot])
```

# New experiment: Henon-Heiles system

```python
# The system
def henon_heiles(t, y, lambda_=1.0):
    x, y, px, py = y
    dxdt = px
    dydt = py
    dpxdt = -x - 2 * lambda_ * x * y
    dpydt = -y - lambda_ * (x**2 - y**2)
    
    return np.array([dxdt, dydt, dpxdt, dpydt])

# Initial conditions
y0 = np.array([0.1, -0.5, 0, 0])  
t_end = 2000
t_span = (0, t_end)  # Time interval
h = 0.05 # Step size

# Solve the system
t_values, y_values = runge_kutta(henon_heiles, t_span, y0, h, Asym36, bsym36, csym36)

# PEP(7,4,6)
t_valuespep, y_valuespep = runge_kutta(henon_heiles, t_span, y0, h*6/5, Apep636, bpep636, cpep636)
# Convert results into arrays
y_values = np.array(y_values)
x_values = y_values[:, 0]
y_values_ = y_values[:, 1]
px_values = y_values[:, 2]
py_values = y_values[:, 3]

y_valuespep = np.array(y_valuespep)
x_valuespep = y_valuespep[:, 0]
y_values_pep = y_valuespep[:, 1]
px_valuespep = y_valuespep[:, 2]
py_valuespep = y_valuespep[:, 3]

# Plotting the trajectory in the (x, y) plane
plt.figure(figsize=(5, 5))
plt.plot(x_values, y_values_, color="blue", lw=0.5, label="PS(5,3,6)")
plt.plot(x_valuespep, y_values_pep, color="red", lw=0.5, label="PEP(6,3,6)")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Hénon-Heiles System")
plt.grid()
plt.legend()
plt.show()
```

```python
# Energy for HH system
def hamiltonian(y, lambda_=1.0):
    x, y, px, py = y
    kinetic = 0.5 * (px**2 + py**2)
    potential = 0.5 * (x**2 + y**2) + lambda_ * (x**2 * y - y**3 / 3)
    return kinetic + potential
# Initial energy
E0 = hamiltonian(y_values[0])
E0pep = hamiltonian(y_valuespep[0])
# Compute energy at each time step
energy_values = [abs(hamiltonian(y)-E0) for y in y_values]
energy_valuespep = [abs(hamiltonian(y)-E0pep) for y in y_valuespep]
# Plot energy over time
plt.figure(figsize=(10, 6))
plt.loglog(t_values, energy_values, label='PS(3,2,4)', color = 'b')
plt.loglog(t_valuespep, energy_valuespep, label='PEP(3,2,4)', color = 'r')
plt.xlabel("t")
plt.ylabel("H(u(t)) − H(u(0))")
plt.title("Hénon-Heiles System: Energy ")
plt.grid()
plt.legend()
plt.savefig("hh_energy324.pdf")
plt.show()
```
